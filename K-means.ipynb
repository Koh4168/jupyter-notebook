```python
from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=300, #サンプルデータ総数
              n_features=2,#特微量の個数
              centers=3,#正規分布の数
              random_state=0)
print(X)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(X[:,0],
                             X[:,1],
                             c='blue',
                             marker='o',
                             s=50)
```

```python
from sklearn.cluster import KMeans
```

```python
# KMenasクラスのインスタンスを作成
km=KMeans(n_clusters=5)
# データをインスタンスに適合させる
km.fit(X)
km_predict = km.predict(X)
print(km_predict)
```

```python
plt.scatter(X[km_predict==0,0],  # グラフのxの値
           X[km_predict==0,1], # グラフのyの値
           s=50,
           c='green',
           label='cluster1')

plt.scatter(X[km_predict ==1,0],
           X[km_predict==1,1],
            s=50,
           c='purple',
           label='cluster2')

plt.scatter(X[km_predict ==2,0],
           X[km_predict==2,1],
            s=50,
           c='red',
           label='cluster3')

plt.scatter(X[km_predict ==3,0],
           X[km_predict==3,1],
            s=50,
           c='blue',
           label='cluster4')

plt.scatter(X[km_predict ==4,0],
           X[km_predict==4,1],
            s=50,
           c='orange',
           label='cluster5')
            
 # クラスターの重心
plt.scatter(km.cluster_centers_[:,0],
           km.cluster_centers_[:,1],
           s=250,
           marker="*",
           c="yellow",
           label="centroid")
```

```python
print('sum pf distance', km.inertia_)
```

```python
# 各クラスター数ごとのSSEを格納しておくための配列を定義
distances=[]
for i in range(1,11): # for文を10回繰り返す
    km=KMeans(n_clusters=i)
    km.fit(X)  # データを学習
    distances.append(km.inertia_ )# SSEを配列に追加
    
plt.plot(range(1,11),distances,marker='.')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Distance')
```

```python
km=KMeans(n_clusters=3)
km.fit(X)
km_predict=km.predict(X)
print(km_predict)
```
